### Parking

```python
# Check how many different values there are in the column parking
house_df['parking'].value_counts()
```

By investigating what different versions of parking mean in Real Estate, I found that these categories are equivalent to "no parking on the property": <br\>
- '0 spaces'
- 'No Data' 
- 'None'
- 'On-street',
- 'Street, None', 
- 'Public Parking', 
- 'On Street, Unassigned', 
- 'No Garage, On Street' <br\>

Therefore, in my definition of *parking* all these parking categories will be assigned a value of zero. The remaining categories will be assigned a value of 1. 

```python
# Transform variable parking from categorical nominal to categorical binary
house_df['parking_binary'] = house_df.parking.apply(lambda x: 0 if x in ['0 spaces','No Data','None','On-street','Street, None', 'Public Parking', 'On Street, Unassigned', 'No Garage, On Street'] else 1)
```
